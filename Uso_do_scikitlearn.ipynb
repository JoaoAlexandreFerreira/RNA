Apender a biblioteca scikit para facilitar problemas de regressão linear

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('data_cov.csv')
y = df['cov']
x = df.drop('cov', axis = 1)
x_treino, x_teste = x[0:86], x[86:]

y_treino, y_teste = y[0:86], y[86:]

escala = StandardScaler()
x_normalizado = escala.fit_transform(x_treino)
print(f"Intervalo de valores        X:{(x_treino)}")
print(f"Intervalor de valores, após normalização X:{(x_normalizado)}")
```

Para realizar o gradiente descendente com o scikit, utilizamos o codigo abaixo

```python
gradiente_descendente_scikit = SGDRegressor(max_iter=350)
gradiente_descendente_scikit.fit(x_normalizado, y_treino)
print(gradiente_descendente_scikit)
print(f"número de interaçoes completas: {gradiente_descendente_scikit.n_iter_}, numero de pesos atualizados: {gradiente_descendente_scikit.t_}")
```

Para pegar o peso e viés que foi armazenado previamente, vamos usar:

```python
b_norm = gradiente_descendente_scikit.intercept_
w_norm = gradiente_descendente_scikit.coef_
print(f"parametros finais: w: {w_norm}, b:{b_norm}")
```

Realizando predições

```python

y_prev_scikit = gradiente_descendente_scikit.predict(x_normalizado)
#predição usando o conhecimento padrão sobre regressão linear
y_prev = np.dot(x_normalizado, w_norm) + b_norm
print(f"A predição com o gradiente descentente é a mesma para scikit e regressão linear? : {(y_prev == y_prev_scikit).all()}")

print(f"Predição utilizandos os dados de entrada:\n{y_prev}" )
print(f"Dados de saída \n{y_treino}")
```
