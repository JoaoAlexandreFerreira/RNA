Baseado no curso de Machine Learning do coursera, esse exemplo trata de uma previsão no preço de um terreno baseado em sua área, baseado no gráfico a seguir:

image.png

Para isso, foram adotados dois pontos, o ponto (1000,300000) e o ponto (2000,500000), para simplificar, os pontos foram divididos por mil, ficando: (1.0, 300) e (2.0, 500).

Implementando em python, sendo x_train os dados de entrada no eixo horizontal e y_train os dados de saída, ambos os valores são listas/vetores.

```python
import numpy as np
import matplotlib.pyplot as plt


x_treino = np.array([1.0, 2.0])
y_treino = np.array([300.0, 500.0])
```

Após isso, precisamos obter o número de exemplos adotados

```python
m = len(x_treino)
m
```

Plotando o gráfico, para visualização os pontos adotados e ter uma noção sobre a forma da reta, que é a regressão linear.

```python
plt.scatter(x_treino, y_treino, marker='x', c='r')
#Titulo do gráfico
plt.title("Preço do terreno")
#Título para o eixo y
plt.ylabel('Preço (em 1000 dol)')
#Título para o eixo x
plt.xlabel('Área (1000 sqft)')
plt.show()
```

Iniciando o modelo de regressão linear, com pesos e viéses

```python
w = 200
b = 100

#para montar a reta, é necessário criar uma função

def reg_linear(x,w,b):
  f_wb = np.zeros(len(x))

  for i in range(len(x)):
    f_wb[i] = w*x[i]+b

  print(f_wb)
  return f_wb

modelo = reg_linear(x_treino,w,b)
```

Agora, realizando uma plotagem do gráfico para ver o comportamento da reta em função dos pontos adotados

```python
plt.plot(x_treino, modelo, c='b',label='Regressão Linear')

plt.scatter(x_treino, y_treino, marker='x', c='r', label = 'Valores fornecidos')
#Titulo do gráfico
plt.title("Preço do terreno")
#Título para o eixo y
plt.ylabel('Preço (em 1000 dol)')
#Título para o eixo x
plt.xlabel('Área (1000 sqft)')
plt.legend()
plt.show()
```

Com uma reta bem ajustada, podemos querer encontrar um ponto no meio do gráfico, para isso:

```python
x_i = 1.5

print("O valor do terreno é de: ",w*x_i + b)
```

O intuito agora é determinar o erro da regressão linear, quando comparado a entrada real da figura um, para isso vamos utilizar o erro médio quadrático. Criando uma função no python para isso.

image.png

```python

def erro_medio_quad(x, y, w, b):

  soma_erro = 0
  for i in range(len(x)):
    f_w = w*x[i] + b
    erro = (f_w - y[i])**2

    soma_erro = soma_erro + erro

  erro_total = (1/2*len(x))*soma_erro

  return erro_total
```
